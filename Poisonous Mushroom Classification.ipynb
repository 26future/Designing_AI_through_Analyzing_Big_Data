```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
mushroom = pd.read_csv("agaricus-lepiota.data", header=None)
mushroom
```

```python
#poisonous = 1, edible = 0
mushroom[0] = np.where(mushroom[0]=='p','1','0')
mushroom
```

```python
mushroom[4] = np.where(mushroom[4]=='f','1','0')
mushroom[6] = np.where(mushroom[6]=='f','1','0')
mushroom[7] = np.where(mushroom[7]=='c','1','0')
mushroom[8] = np.where(mushroom[8]=='b','1','0')
mushroom[10] = np.where(mushroom[10]=='e','1','0')
```

```python
mushroom
```

```python
mushroom.drop(16, axis=1, inplace=True)
```

```python
mushroom
```

```python
mushroom_dummies = pd.get_dummies(data=mushroom, columns=[1,2,3,5,9,11,12,13,14,15,17,18,19,20,21,22], prefix=[1,2,3,5,9,11,12,13,14,15,17,18,19,20,21,22])
mushroom_dummies
```

```python
xdata = mushroom_dummies.iloc[:,1:]
xdata
ydata = mushroom_dummies.iloc[:,0]
ydata
```

```python
xdata.shape #(8124, 111)
ydata.shape #(8124,)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xdata, ydata, random_state=42, train_size=0.7)
```

## Logistic classification

```python
res = LogisticRegression().fit(xtrain, ytrain)
```

```python
print('training set score: {:.3f}'.format(res.score(xtrain,ytrain)))
print('test set score: {:.3f}'.format(res.score(xtest,ytest)))
```

## knn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
res = clf.fit(xtrain, ytrain)
```

```python
clf.predict(xtest)
```

```python
print("training set score: {:.3f}".format(res.score(xtrain,ytrain)))
print("test set score: {:.3f}".format(res.score(xtest,ytest)))
```
