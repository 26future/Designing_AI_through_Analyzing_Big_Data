```python
#여러 개의 입력값을 전달 받는 함수
#def 함수명(*매개변수):
#    수행문장

def add_many(*args):
    res=0
    for i in args:
        res+=i
    return res
  

print(add_many(1,2,3))
#res=add_many(1,2,3)
#print(res)

res = add_many(1,2,3,4,5,6)
print(res)
```

```python
def add_mul(ch, *args):
    if ch == 'add':
        res=0
        for i in args:
            res+=i
    
    elif ch == 'mul':
        res=1
        for i in args:
            res*=i  
    return res  
                
print(add_mul('add', 1,2,3))

print(add_mul('mul', 1,2,3))
```

```python
#람다표현식에 조건부 표현식 적용
a = list(range(1,11))
#a에 저장된 요소값들 중에서 3의 배수는 100을 더해서 출력
#[1, 2, 103, 4, 5, 106, 7, 8, 109, 10]


#for i in range(1,11):
#    if a[i]%3 == 0:
#        a[i]+=100
        #a[i] = str(a[i])
#print(a)
        
#위 문장을 람다표현식으로 바꿔보자
#map(람다표현식,a)
print(list(map(lambda x:str(x) if x%3==0 else x,a)))
#lambda 매개변수:식1 if조건식 else 식2 
#람다식에 if문이 있으면 반드시 else를 사용해야 함
#조건식1이 참이면 식1 실행, 거짓이면 식2 실행. 
```

```python
#람다식에는 elif를 사용 못함
#식1 if 조건식1 else 식2 if 조건식2 else 식3  #조건식1이 참이면 식1 실행, 조건식2가 참이면 식2 실행. 그렇지 않으면 식3 실행

#if 조건식:
#    식1
#elif 조건식:
#    식2
#else:
#    식3

#a에 저장되어 있는 데이터에 대해, 1은 문자열로 변환. 2는 실수로 변환. 나머지 3이상은 모두 10을 더해서 출력
a = range(1,11)

for i in a:
    if a[i]==1:
        print(str(a[i]))
    elif a[i]==2:
        print(float(a[i]))
    else:
        s = (a[i]+10)
        print(s)        
```

```python
#지역변수: 함수 내부에서 변수 선언. 함수 내부에서만 접근 가능
#함수 외부에서는 접근 못함. 지역범위
def f2():
    x2=2
    print(x2)
f2()
print(x2)
```

```python
x=10  #전역변수
def f3():
    #global : 함수 내부에서 전역변수 값을 변경하고자 하는 경우
    #global 뒤에 전역 변수 이름을 지정
    global x
    x=20  #전역변수
    print(x) #전역변수값 출력
f3()
print(x) #전역변수값 출력 (20)
```

```python
def f4():
    global x4  #x4 변수를 전역변수화
    x4=20
    print(x4)
f4()
print(x4)  #전역변수 x4를 실행
```

```python
#함수 내부에 함수 정의
"""
def 함수1
    문장
    def 함수2()
    문장
"""
def outer():
    msg='hello'  #전역변수
    def inner():
        print(msg)
    inner()
outer()
```

```python
#전체줄을 읽어서 출력
f = open("c:/da/newFile.txt",'r')
while True:
    line = f.readline() #더이상 읽어들일 줄이 없으면 none 출력
    if not line: break                
    print(line)
f.close()
```
