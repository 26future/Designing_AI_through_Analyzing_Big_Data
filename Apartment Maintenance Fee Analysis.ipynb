```python
import pandas as pd
import numpy as np
```

```python
df0 = pd.DataFrame(['일반관리비','청소비','경비비','소독비','승강기유지비','지능형홈네트워크설비유지비','수선유지비','위탁관리수수료','난방공용','급탕공용','가스공용','전기공용','수도공용'])
month = ['01','02','03','04','05','06','07','08','09','10','11','12']
year = ['2015','2016','2017','2018','2019']
for m in month:
    for y in year:
        file = pd.read_excel('관리비데이터/'+y+m+' 가양도시개발공사8단지(임대).xls', header=[2])
        file = file.loc[[1,32,33,34,35,36,37,42,45,48,51,54,57],]
        df = file['우리단지총액']
        df = pd.DataFrame(df)
        df = df.reset_index(drop=True)
        df.columns = [y+m]
        df0 = df0.join(df)
        
df0
```

```python
#지능형홈네트워크설비유지비,위탁관리수수료,급탕공용,가스공용 제외
df_y = df_y[df_y!=0].dropna()
df_y
```

```python
#연도별 데이터
for m in month:
    li=[]
    for y in year:
        li.append(y+m)
        df_y[y]=df_y[li].sum(axis=1)
        df_y
df_y
```

```python
#분기별 데이터
df_q = df_q[['1st','2nd','3rd','4th']]
df_quarter = df_q/15
df_quarter = df_quarter.T
df_quarter
```

```python
#시각화
df_quarter.plot(figsize=(10,6))
plt.title('quarterly cost')
plt.xlabel('quarter')
plt.ylabel('cost')
plt.legend(loc='right', bbox_to_anchor=(1.5, 0.5))
```

```python
#4. 관리비 항목의 계절별 변화 패턴
#1,2,11,12 : winter, 3,4,5 : spring, 6,7,8: summer, 9,10:fall
df_s = df_m.copy()
df_s
```

```python
#계절별 데이터
df_s['winter'] = df_s[['01','02','11','12']].sum(axis=1)
df_s['winter'] = df_s['winter']/20
df_s['spring'] = df_s[['03','04','05']].sum(axis=1)
df_s['spring'] = df_s['spring']/15
df_s['summer'] = df_s[['06','07','08']].sum(axis=1)
df_s['summer'] = df_s['summer']/15
df_s['fall'] = df_s[['06','07','08']].sum(axis=1)
df_s['fall'] = df_s['fall']/15
df_s
```

```python
df_s = df_s[['spring','summer','fall','winter']]
df_season = df_s.T
df_season
```

```python
#시각화
df_season.plot(figsize=(10,6))
plt.title('seasonal cost')
plt.xlabel('season')
plt.ylabel('cost')
plt.legend(loc='right', bbox_to_anchor=(1.5, 0.5))
```

## 2020~2024년까지 5년 동안의 관리비 예측

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
idx = list(df0.columns)
del(idx[0])
```

```python
df_x = pd.DataFrame(idx).set_index(0)
df_x
```

```python
df_x['year'] = None
df_x['month'] = None
df_x['quarter'] = None
df_x['season'] = None
df_x['monthly_avg'] = None
df_x
```

```python
year_new = ['2020','2021','2022','2023','2024']
for y in year_new:
    for m in month:
        df_x.loc[y+m,] = None
df_x
```

```python
total_year = list(range(2015,2025))

for y in total_year:
    df_x.loc[[i for i in df_x.index if str(y) in str(i)],'year'] = y
df_x
```

```python
for m in month:
    df_x.loc[[i for i in df_x.index if str(i).endswith(m)],'month'] = m

df_x
```

```python
quarter1 = ['01','02','03']
quarter2 = ['04','05','06']
quarter3 = ['07','08','09']
quarter4 = ['10','11','12']
for q1 in quarter1:
    df_x.loc[[i for i in df_x.index if str(i).endswith(q1)],'quarter'] = '1'

for q2 in quarter2:
    df_x.loc[[i for i in df_x.index if str(i).endswith(q2)],'quarter'] = '2'
    
for q3 in quarter3:
    df_x.loc[[i for i in df_x.index if str(i).endswith(q3)],'quarter'] = '3'
    
for q4 in quarter4:
    df_x.loc[[i for i in df_x.index if str(i).endswith(q4)],'quarter'] = '4'

df_x
```

```python
#1,2,11,12 : 겨울, 3,4,5 : 봄, 6,7,8: 여름, 9,10:가을
winter = ['01','02','11','12']
spring = ['03','04','05']
summer = ['06','07','08']
fall = ['09','10']

for w in winter:
    df_x.loc[[i for i in df_x.index if str(i).endswith(w)],'season'] = 'winter'

for sp in spring:
    df_x.loc[[i for i in df_x.index if str(i).endswith(sp)],'season'] = 'spring'

for su in summer:
    df_x.loc[[i for i in df_x.index if str(i).endswith(su)],'season'] = 'summer'

for f in fall:
    df_x.loc[[i for i in df_x.index if str(i).endswith(f)],'season'] = 'fall'
    
df_x
```

```python
df_month['Total'] = df_month.sum(axis=1)
df_month
```

```python
for m in month:
    df_x.loc[[i for i in df_x.index if str(i).endswith(m)],'monthly_avg'] = df_month.loc[m,'Total']
df_x
```

```python
total_cost = df_cost.sum()
total_cost = pd.DataFrame(total_cost)
total_cost.drop(['Total','percentage'], inplace=True)
total_cost = total_cost.rename({0:'total_cost'}, axis='columns')
total_cost
```

```python
df_x = df_x.join(total_cost)
```

```python
dfWithDummies = pd.get_dummies(data=df_x, columns=['year','month','quarter','season'],prefix=['year','month','quarter','season'])
dfWithDummies
```

```python
data_Y = dfWithDummies.iloc[-60:]
data_X = dfWithDummies.iloc[:60]
```

```python
data_Y.drop('total_cost', axis=1, inplace=True)
data_Y
```

## KNN 회귀 분석

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_X,data_Y, random_state=42) #train/test data 나누기
reg = KNeighborsRegressor(n_neighbors=3) #neighbor=3으로 객체 생성
reg.fit(x_train,y_train) #모델 학습
```

```python
print("테스트 세트 예측:", reg.predict(x_test))
```
