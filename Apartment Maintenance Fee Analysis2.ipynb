```python
import pandas as pd
```

```python
"""
2. 아파트 관리비 분석
(1) 관리비 구성
- 공용관리비=일반관리비+청소비+경비비+소독비+승강기유지비+수선유지비
- 일반관리비=인건비+제세공과금
- 제세공과금=통신료+제세공과금등
(2) 공용관리비 항목에 대해 아래와 같은 분석을 실시하시오.
1) 일반관리비, 청소비, 경비비, 소독비, 승강기유지비, 수선유지비 각 항목에 대해 공용관리비에서 차지하는 비중의 변화가 어떠한지 연도별로 분석하시오.(시각화)
2) 공용관리비를 줄이고자 한다고 가정했을때, 일반관리비(인건비 제외), 청소비, 경비비, 소독비, 승강기유지비, 수선유지비 항목중 증가되는 비율이 가장 높은 항목(비용을 줄여야 하는 항목)은 무엇인가?(시각화)
3) 특정 연도에  과다 청구된 해가 있는지 조사하시오.(시각화)
(3) 일반관리비 항목에 대해 아래와 같은 분석을 실시하시오.
1) 인건비, 제사무비, 제세공과금, 피복비, 교육훈련비, 차량유지비, 그밖의부대비용 각 항목에 대해 일반관리비에서 차지하는 비중의 변화가 어떠한지 연도별로 분석하시오.(시각화)
2) 특정 연도에  과다 청구된 해가 있는지 조사하시오.(시각화)
"""
```

```python
li=[]
year = ['2012','2013','2014','2015','2016','2017','2018','2019']
for y in year:
    df = pd.read_excel("아파트관리비분석용 데이터셋/아파트관리비분석용 데이터셋/"+y+"년.xls")
    li.append(df.set_index("분류"))
print(li)
```

```python
df_2012 = li[0]
df_2013 = li[1]
df_2014 = li[2]
df_2015 = li[3]
df_2016 = li[4]
df_2017 = li[5]
df_2018 = li[6]
df_2019 = li[7]
```

## 공용관리비

```python
# 1-1. 각 항목(일반관리비, 청소비, 경비비, 소독비, 승강기유지비, 수선유지비)이 
#공용관리비에서 차지하는 비중의 변화 => 연도별 분석
```

```python
total_li=[]
for df in li:
    df_cost = df.loc[['일반관리비', '청소비', '경비비', '소독비', '승강기유지비', '수선유지비'],'합계']
    df_cost = pd.DataFrame(df_cost).astype(float)
    df = (df_cost/df_cost.sum())*100
    total_li.append(df)
print(total_li)
```

```python
total_2012 = total_li[0]
total_2013 = total_li[1]
total_2014 = total_li[2]
total_2015 = total_li[3]
total_2016 = total_li[4]
total_2017 = total_li[5]
total_2018 = total_li[6]
total_2019 = total_li[7]
```

```python
# 1-2. 일반관리비(인건비 제외), 청소비, 경비비, 소독비, 승강기유지비, 수선유지비 항목중 
# 증가되는 비율이 가장 높은 항목 (시각화)
```

```python
increase_df = pd.DataFrame([], index=['일반관리비', '청소비', '경비비', '소독비', '승강기유지비', '수선유지비'])
for y in year:
    df = pd.read_excel("아파트관리비분석용 데이터셋/아파트관리비분석용 데이터셋/"+y+"년.xls")
    df = df.set_index('분류')
    df = pd.DataFrame(df.loc[['일반관리비', '인건비', '청소비', '경비비', '소독비', '승강기유지비', '수선유지비'],'합계']).astype(float)
    df.loc['일반관리비','합계'] = int(df.loc['일반관리비','합계'])-int(df.loc['인건비','합계'])
    df.drop('인건비', inplace=True) #일반관리비에서 인건비 제외
    df = df.rename({'합계': y}, axis='columns')
    increase_df = increase_df.join(df)
display(increase_df)
```

```python
#시각화

plt.plot(increase_df.loc['일반관리비',:])
plt.plot(increase_df.loc['청소비',:])
plt.plot(increase_df.loc['경비비',:])
plt.plot(increase_df.loc['소독비',:])
plt.plot(increase_df.loc['승강기유지비',:])
plt.plot(increase_df.loc['수선유지비',:])

plt.xlabel("year")
plt.ylabel("total cost")
plt.legend(loc='right', bbox_to_anchor=(1.35, 0.5))
```

```python
cost_df = pd.DataFrame([], index=['인건비', '제사무비', '제세공과금', '피복비', '교육훈련비', '차량유지비','그밖의부대비용'])
for y in year:
    df = pd.read_excel("아파트관리비분석용 데이터셋/아파트관리비분석용 데이터셋/"+y+"년.xls")
    df = df.set_index('분류')
    df = pd.DataFrame(df.loc[['인건비', '제사무비', '제세공과금', '피복비', '교육훈련비', '차량유지비','그밖의부대비용'],'합계']).astype(float)
    df.rename({'합계':y}, axis='columns')
    df = df.rename({'합계': y}, axis='columns')
    cost_df = cost_df.join(df)
display(cost_df)
```
