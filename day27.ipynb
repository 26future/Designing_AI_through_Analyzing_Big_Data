```python
from lxml import etree
```

```python
tree = etree.parse('sample.xml') #xml파일을 tree구조로 데이터를 읽음
root = tree.getroot() #root tag를 가져옴
```

```python
members=[]
albums=[]
kids = root.getchildren() #자식 tag를 가져옴
for child in kids:
#     print(child.tag) #tag이름 출력
#     print(child.text) #안에 text를 가지고 있으면, text내용 출력
    if child.tag == 'name':
        gname = child.text
    elif child.tag == 'members':
        for xmember in child:
            members.append(xmember.text)
    elif child.tag == 'albums':
        for xalbum in child:
            albums.append([xalbum.get('order'),xalbum.text]) #특정 속성 #리스트 안의 리스트
```

```python
print('걸그룹: %s' %gname)
print('멤버:', end='')
for i,m in enumerate(members): #번호,이름
    print(m," ",end='')
print()
for album in albums:
    print("%s:%s" %(album[0],album[1]))
```

```python
#JSON 파일 읽어오기
import json
```

```python
path = 'usagov_bitly_data.txt'

records = [json.loads(line) for line in open(path, encoding='utf-8')]
records
records[0]
records[0]['tz']

time_zones = [rec['tz'] for rec in records if 'tz' in rec]
#tz컬럼이 있는 경우에만 time_zones에 저장
time_zones
```

```python
def get_counts(seq):
    counts={} #딕셔너리
    for x in seq:
        if x in counts:
            counts[x]+=1
        else:
            counts[x] = 1
            #counts['America/New_York']=1
    return counts
counts = get_counts(time_zones)
```

```python
counts
counts['America/New_York']
len(time_zones)
```

```python
from collections import defaultdict
```

```python
def get_counts2(seq):
    counts = defaultdict(int) #값을 0으로 초기화한 딕셔너리
    for x in seq:
        counts[x]+=1
    return counts

counts = get_counts2(time_zones)
counts
```

```python
time_zones
counts
#가장 많이 등장하는 상위 10개의 표준 시간대를 출력
# [(1251, 'A/N'),
#  (521, ' ')]
def top_counts(count_dict, n=10): #n=10 default, 새로운 인수를 받으면, 전달받은 값으로 변경됨
    pairs = [(count,tz) for tz, count in  count_dict.items()]
    pairs.sort()
    return pairs[-n:] #뒤에서부터 10개
print(top_counts(counts))
    

# def top10(counts):
#     top_counts = sorted(counts.items(), key=lambda x:x[1], reverse=True)
#     return top_counts[:10]
# top10(counts)
```
