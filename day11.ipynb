```python
from selenium import webdriver
url = "http://www.naver.com"
#phantomjs 드라이버 추출
browser = webdriver.PhantomJS()
#3초 대기
browser.implicitly_wait(3)
#url 읽기
browser.get(url)
#화면 캡쳐
browser.save_screenshot("myshot.png")
#브라우저 종료
browser.quit()
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://stackoverflow.com/questions/tagged/python"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

links=[]
for link in soup.select('div.question-summary h3 a'):
    links.append(link.attrs['href'])
for link in links:
    url = "https://stackoverflow.com" + link
    response = requests.get(url)
    text = response.text
    soup = Beautifulsoup(text, 'html.parser')
    print("Question\n")
    for x in soup.select('div.postcell div.post=text p'):
        print(x.text)
    print("\nAnswer\n")
    for x in soup.select('div.answercell div.post=text p'):
        print(x.text)
    print("="*40, "\n")
    
    
#     print(soup)
#     print("Question\n")
    
#     print("\nAnswer\n")
    
#     print("="*50,'\n')
```

```python
#1-10페이지 질문, 링크

import requests
from bs4 import BeautifulSoup

for page in range(1,92,10):
    url = "https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&c_id=&c_name=&sm=tab_pge&kin_start="+str(page)
    response = requests.get(url)
    response
    soup = BeautifulSoup(response.text, 'html.parser')
    for number in range(1,11):
        questions = soup.select_one("#elThumbnailResultArea > li:nth-child("+str(number)+") > dl > dt > a").text
        print("질문:"+questions)
        link = soup.select_one("#elThumbnailResultArea > li:nth-child("+str(number)+") > dl > dt > a").get('href')
        print(link+"\n")
```

```python
#1-2페이지에 있는 질문/답변 모두 추출

import requests
from bs4 import BeautifulSoup


for page in range(1,12,10):
    url = "https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%ED%8C%8C%EC%9D%B4%EC%8D%AC&c_id=&c_name=&sm=tab_pge&kin_start="+str(page)
    response = requests.get(url)
    response
    soup = BeautifulSoup(response.text, 'html.parser')
    
    links=[]
    for number in range(1,11):
        questions = soup.select_one("#elThumbnailResultArea > li:nth-child("+str(number)+") > dl > dt > a").text
        print("질문:"+questions+"\n")
        link = soup.select_one("#elThumbnailResultArea > li:nth-child("+str(number)+") > dl > dt > a").get('href')
        print(link)
        links.append(link)

    for link in links:
        response = requests.get(link)
        response
        soup = BeautifulSoup(response.text, 'html.parser')
        answers = soup.select_one("#answer_1 > div._endContents.c-heading-answer__content > div._endContentsText.c-heading-answer__content-user").text
        print("답변:"+answers.lstrip())
```
